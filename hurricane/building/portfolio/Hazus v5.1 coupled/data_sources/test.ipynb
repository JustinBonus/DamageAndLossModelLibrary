```python
import json
import pandas as pd

with open(
    '/Users/adamzs/Repos/PBE/DamageAndLossModelLibrary/hurricane/building/component/SimCenter Wind Component Library/fragility_header.json',
    'r',
) as f:
    meta = json.load(f)

input_data = pd.read_csv(
    '/Users/adamzs/Repos/PBE/DamageAndLossModelLibrary/hurricane/building/component/SimCenter Wind Component Library/fragility_source.csv'
)
```

```python
fragility_cols = [
    'ID',
    'Incomplete',
    'Demand-Type',
    'Demand-Unit',
    'Demand-Offset',
    'Demand-Directional',
    'LS1-Family',
    'LS1-Theta_0',
    'LS1-Theta_1',
]

fragility_data = (
    input_data[fragility_cols].set_index('ID').sort_index().convert_dtypes()
)

fragility_data.to_csv(
    '/Users/adamzs/Repos/PBE/DamageAndLossModelLibrary/hurricane/building/component/SimCenter Wind Component Library/fragility.csv'
)

fragility_data
```

```python
meta_plus = {}

for row_id, row in input_data.iterrows():
    component_meta = {
        'Description': row['Meta-Description'],
        'Comments': row['Meta-Comments'],
        'SuggestedComponentBlockSize': row['Meta-SuggestedComponentBlockSize'],
        'RoudUpToIntegerQuantity': row['Meta-RoundUpToIntegerQuantity'],
        'Reference': [ref.strip() for ref in row['Meta-Reference'].split(',')],
        'LimitStates': {
            'LS1': {
                'DS1': {'Description': row['Meta-LimitStates-LS1-DS1-Description']}
            }
        },
    }

    for ref in component_meta['Reference']:
        if ref not in meta['References']:
            print(f'Error, missing reference: {ref}')

    meta_plus[row['ID']] = component_meta
```

```python
meta.update(meta_plus)
```

```python
with open(
    '/Users/adamzs/Repos/PBE/DamageAndLossModelLibrary/hurricane/building/component/SimCenter Wind Component Library/fragility.json',
    'w',
) as f:
    json.dump(meta, f, indent=2)
```
